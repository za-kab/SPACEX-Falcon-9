```python
import folium
import pandas as pd

# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
spacex_df = pd.read_csv(URL)
spacex_df.head(5)
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

spacex_df.tail()
```

```python
marker_colour = []

for c in spacex_df['class']:
    if c == 1:
        marker_colour.append('green')
    else: 
        marker_colour.append('red')

spacex_df['marker_colour'] = marker_colour
spacex_df.head(20)
```

```python
# launch sites cooordinate
CCAFSLC_coord = launch_sites_df.loc[0][1:3].to_list()
CCAFSSLC_coord = launch_sites_df.loc[1][1:3].to_list()
KSC_coord = launch_sites_df.loc[2][1:3].to_list()
VAFB_coord = launch_sites_df.loc[3][1:3].to_list()
```

```python
for index, record in spacex_df.iterrows():
    print(record['Lat'])
    print(record['Long'])
    print(record['marker_colour'])
```

```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
```

```python
# Create a circle at each launch site
NASA_circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
CCAFSLC_circle = folium.Circle(CCAFSLC_coord, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('Cape Canaveral LC-40'))
CCAFSSLC_circle = folium.Circle(CCAFSSLC_coord, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('Cape Canaveral SLC-40'))
KSC_circle = folium.Circle(KSC_coord, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('Kennedy Space Center LC-39A'))
VAFB_circle = folium.Circle(VAFB_coord, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('Vandenberg SLC-4E'))

# Create a marker at each launch site
NASA_marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC'
        )
    )
CCAFSLC_marker = folium.map.Marker(CCAFSLC_coord,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'CCAFS LC-40'
        )
    )
CCAFSSLC_marker = folium.map.Marker(CCAFSSLC_coord,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'CCAFS SLC-40'
        )
    )
KSC_marker = folium.map.Marker(KSC_coord,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "KSC LC-39A"
        )
    )
VAFB_marker = folium.map.Marker(VAFB_coord,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'VAFB SLC-4E'
        )
    )

site_map.add_child(NASA_circle)
site_map.add_child(CCAFSLC_circle)
site_map.add_child(CCAFSSLC_circle)
site_map.add_child(KSC_circle)
site_map.add_child(VAFB_circle)

site_map.add_child(NASA_marker)
site_map.add_child(CCAFSLC_marker)
site_map.add_child(CCAFSSLC_marker)
site_map.add_child(KSC_marker)
site_map.add_child(VAFB_marker)

marker_cluster = MarkerCluster()

# Add the MarkerCluster to the site_map
site_map.add_child(marker_cluster)

# Loop through each row in spacex_df to create and add markers to the MarkerCluster
for index, record in spacex_df.iterrows():
    # Extract the data you need
    lat = record['Lat']
    lon = record['Long']
    marker_color = record['marker_colour']  # Extract the marker color

    # Create a Marker object
    marker = folium.Marker(
        location=[lat, lon],
        icon=folium.Icon(color='white', icon_color=marker_color)  # Customize icon
    )

    # Add the Marker to the MarkerCluster
    marker_cluster.add_child(marker)

```

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
# function to calculate distatnce
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
